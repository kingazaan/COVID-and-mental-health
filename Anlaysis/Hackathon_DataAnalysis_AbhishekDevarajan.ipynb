# Question 1: Which countries are the most negative, and which are the most positive

```python
country_sentiment = df[['country','tone_numeric']].groupby('country').sum()
country_sentiment_top10 = country_sentiment.sort_values('tone_numeric',ascending=False).reset_index()[:10]
country_sentiment_bottom10 = country_sentiment.sort_values('tone_numeric',ascending=True).reset_index()[:10]

fig,ax = plt.subplots(1,2,figsize=(12,10))
sns.barplot(country_sentiment_top10['country'],country_sentiment_top10['tone_numeric'],ax = ax[0])
sns.barplot(country_sentiment_bottom10['country'],country_sentiment_bottom10['tone_numeric'], ax = ax[1])
ax[0].set_xlabel('Country')
ax[0].set_ylabel('Net Sentiment')
ax[0].set_title('Top 10 Most Positive Countries')
ax[1].set_xlabel('Country')
ax[1].set_ylabel('Net Sentiment')
ax[1].set_title('Top 10 Most Negative Countries')
plt.show()
```

**Insights:** The graphs show that the most negative country is Great Britain, while the most positive is India. Both of these countries have a much larger magnitude of sentiment than the others in the top and bottom 10. I suspect that this might have something to do with the larger english speaking populations in those countries

## Question 1b: What does the distribution of net sentiment look like?

```python
fig = plt.figure(figsize=(10,10))
sns.distplot(country_sentiment['tone_numeric'])
plt.show()
```

**Insights:** Most countries are pretty much neutral in terms of sentiment. This could either be because of equal numbers of positive and negative posts, or because of a lack of positive and negative posts
